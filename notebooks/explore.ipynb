```python
import pandas as pd
```

Loading raw data

```python
path = '../data-raw/ncr_ride_bookings.csv'
```

```python
data_raw = pd.read_csv(path) 
data_raw.head()
```

Exploring the data

```python
data_raw.info()
```

Treating NaN values

In some columns, changing null values to 0

```python
data_fillna = data_raw.copy()

data_fillna['Avg CTAT'] = data_fillna['Avg CTAT'].fillna(0)
data_fillna['Avg VTAT'] = data_fillna['Avg CTAT'].fillna(0)
data_fillna['Cancelled Rides by Customer'] = data_fillna['Cancelled Rides by Customer'].fillna(0)
data_fillna['Cancelled Rides by Driver'] = data_fillna['Cancelled Rides by Driver'].fillna(0)
data_fillna['Incomplete Rides'] = data_fillna['Incomplete Rides'].fillna(0)


data_fillna
```

Adding new column do representing the total time of waiting by adding AVG PTAT + AVG CTAT

```python
data_fillna['AVG Waiting Time'] = data_fillna['Avg CTAT'] + data_fillna['Avg VTAT']

data_fillna.info()
```

Transforming the date column

```python
data_fillna['Date'] = pd.to_datetime(data_fillna['Date'])
```

```python
data_fillna.info()
```

```python
data_fillna.head()
```

Transforming float into int when float is not necessary

```python
data_fillna['Cancelled Rides by Customer'] = data_fillna['Cancelled Rides by Customer'].astype(int)
data_fillna['Cancelled Rides by Driver'] = data_fillna['Cancelled Rides by Customer'].astype(int)
data_fillna['Incomplete Rides'] = data_fillna['Incomplete Rides'].astype(int)

data_fillna.head()
```
